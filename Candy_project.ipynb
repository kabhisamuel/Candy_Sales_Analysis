```python
#importing Libraries

import pandas as pd
import numpy as np

# Loading Sales,Products,Factories,Targets,Data dictionariess datasets

sales = pd.read_csv('Candy_Sales.csv')
products = pd.read_csv('Candy_Products.csv')
factories = pd.read_csv('Candy_Factories.csv')
targets = pd.read_csv('Candy_Targets.csv')
data_dict = pd.read_csv('candy_distributor_data_dictionary.csv')
```

```python
# Converting dates in sales to change Date in string format to datetime datatype.
sales['Order Date'] = pd.to_datetime(sales['Order Date'])
sales['Ship Date'] = pd.to_datetime(sales['Ship Date'])
print(sales['Order Date'])
print(sales['Ship Date'])
```

```python
# Adding derived columns like Year,Profit Margin.
sales['Year'] = sales['Order Date'].dt.year
sales['Profit Margin'] = (sales['Gross Profit'] / sales['Sales']) * 100
```

```python
# Checking for duplicates and missing values (none found).
sales.drop_duplicates(inplace=True)  
print(sales.isnull().sum())  #Returns 0 if null values are not found.
factories.drop_duplicates(inplace=True)
print(factories.isnull().sum())
targets.drop_duplicates(inplace=True)
print(targets.isnull().sum())
products.drop_duplicates(inplace=True)
print(products.isnull().sum())

```

```python
# Mergeing datasets sales,targets,products,factories for analysis.
# Creating Candy_master_table 
Candy_master_table= pd.merge(sales, products, on=['Product ID', 'Product Name', 'Division'], how='left')
Candy_master_table = pd.merge(Candy_master_table, factories, left_on='Factory', right_on='Factory', how='left')
Candy_master_table= pd.merge(Candy_master_table, targets, on='Division', how='left')  


```

```python
#Quick Preview of Candy_master_table.
print(Candy_master_table.head())
print(Candy_master_table.describe())
print(Candy_master_table.info())
```

```python
#Removing null values in Candy_master_table.
Candy_master_table.dropna(inplace=True)   
```

```python
#Creating new Master table csv file and reading the csv file.

Candy_master_table.to_csv('Candy_master_dataset.csv', index=False)
df=pd.read_csv('Candy_master_dataset.csv')
```
